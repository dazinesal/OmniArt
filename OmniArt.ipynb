# Data Cleaning

## Removing Irrelevant Data

### Null Values

```python
def vis_null(ds):
    ds_null = ds.isnull()
    fig, ax = plt.subplots(figsize=(10,10)) 
    sns.heatmap(~ds_null, cbar=False, yticklabels=False, ax=ax)
```

```python
vis_null(omnia)
```

```python
print('Before: '+str(omnia.shape))
omni = omnia.dropna(axis=0, how='any')
print('After: '+str(omni.shape))
```

```python
vis_null(omni)
```

### Duplicates

General Duplicates

```python
print('Before: '+str(omni.shape))
omn = omni.drop_duplicates()
print('After: '+str(omn.shape))
```

Duplicates by Omni ID

```python
print('Before: '+str(omn.shape))
om = omn.drop_duplicates('omni_id', keep='last')
print('After: '+str(om.shape))
```

## Transforming Data

### Transforming Columns

```python
om.columns
```

```python
cols = om.columns
newcols={}
for i in cols:
    newi=i.strip()
    newi=newi.replace("_"," ")
    newi = newi.title()
    newcols[i]=newi
om.rename(columns = newcols, inplace = True)
newcols
```

### Transforming Rows

```python
om.info()
```

#### Numerical Type

Century

```python
om['Century']
```

```python
a = [str(i) for i in om["Century"]]

cent = []
for i in a:
    p = re.compile("(\d+)")
    result = p.search(i)
    if result == None:
        cent.append(0)
    else:
        cent.append(float(result.group(1)))
    
cent
```

```python
cent.count(0)
```

```python
om['Century']= cent
```

Creation Year

```python
om['Creation Year']
```

```python
a = [str(i) for i in om["Creation Year"]]

year = []
for i in a:
    p = re.compile("(\d+)")
    result = p.search(i)
    if result == None:
        year.append(0)
    else:
        if (float(result.group(1))>2021):
            year.append(0)
        else:
            year.append(float(result.group(1)))

year
```

```python
year.count(0)
```

```python
om['Creation Year']= year
```

#### String Type

Color Palette

```python
om["Color Pallete"]
```

```python
a = [str(i) for i in om["Color Pallete"]]

pal = []
for i in a:
    r=i.strip("[").strip("]").replace('u\'','').replace('\'','').replace(',','')
    pal.append(r)
pal
```

```python
om['Color Pallete']= pal
```

# Color Classification Using HSV

```python
def hex_to_rgb(hex):
    hex = str(hex)
    hex = hex.lstrip('#')
    hlen = len(hex)
    return tuple(int(hex[i:i + hlen // 3], 16) for i in range(0, hlen, hlen // 3))

def rgb_to_hsv(rgb):
    return colorsys.rgb_to_hsv(rgb[0]/255,rgb[1]/255,rgb[2]/255)

def hex_to_hsv(hex):
    return rgb_to_hsv(hex_to_rgb(hex))
```

```python
def color_cat(hex):
    hsv=hex_to_hsv(hex)
    hue=hsv[0]*360
    sat=hsv[1]*360
    val=hsv[2]*360
    if val<108: return 'Black'
    elif sat<36 and val<342: return 'Gray'
    elif sat<36: return 'White'
    elif hue<30: return 'Red'
    elif hue<45: return 'Orange'
    elif hue<65: return 'Yellow'
    elif hue<170: return 'Green'
    elif hue<190: return 'Turquoise'
    elif hue<250: return 'Blue'
    elif hue<280: return 'Purple'
    elif hue<330: return 'Pink'
    elif hue<=360: return 'Red'
```

```python
color = color_cat('#116E3E')
print(color)
```

```python
om['Color Category'] = om['Dominant Color'].apply(color_cat)
```
